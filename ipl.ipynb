```python
mdata = pd.read_csv(r'D:\Projects\Ipl\Ipl-Kaggle-Dataset-Analysis\matches.csv')
mdata.replace(['Mumbai Indians','Kolkata Knight Riders','Royal Challengers Bangalore','Deccan Chargers','Chennai Super Kings',
                 'Rajasthan Royals','Delhi Daredevils','Gujarat Lions','Kings XI Punjab',
                 'Sunrisers Hyderabad','Rising Pune Supergiants','Kochi Tuskers Kerala','Pune Warriors','Rising Pune Supergiant']
                ,['MI','KKR','RCB','DC','CSK','RR','DD','GL','KXIP','SRH','RPS','KTK','PW','RPS'],inplace=True)
```

```python
sub_data = mdata.iloc[:, 4:7]
```

```python
sub_data.shape
```

```python
team = sub_data['toss_winner'].value_counts().idxmax()
is_team = (sub_data['team1'] == team) | (sub_data['team2'] == team)
team_match_data = sub_data[is_team]
```

```python
matches = team_match_data['toss_winner'].value_counts().sum()
won = team_match_data['toss_winner'].value_counts().max()
percentage = (won/matches)*100
```

```python
print(team,"has won",won,"tosses out of",matches,"making it the highest toss wins in the league at "
                                                 "{0:0.2f}".format(percentage),"%")
```

```python
teams = sub_data['toss_winner'].unique()
print (teams)
```

```python
percs = []
for team in teams:
    is_team = (sub_data['team1'] == team) | (sub_data['team2'] == team)
    team_match_data = sub_data[is_team]
    matches = team_match_data['toss_winner'].value_counts().sum()
    won = team_match_data['toss_winner'].value_counts().max()
    percentage = (won/matches)*100
    percs.append(won)
```

```python
import matplotlib.pyplot as plt
fig = plt.gcf()
fig.set_size_inches(15,7)
plt.bar(teams, percs, color='grey')
plt.title('Toss Wins')
plt.xlabel('Teams')
plt.ylabel('Tosses Won')
plt.show()
```

#### Analyzing the toss decisions at each venue

```python
total_toss = mdata.shape[0]
sub_data = mdata.loc[:, ['toss_winner','winner']]
toss_wins_match = sub_data['toss_winner'] == sub_data['winner']
sub_data = sub_data[toss_wins_match]
wins_tosses = sub_data.shape[0]
percentage = (wins_tosses/total_toss)*100
```

```python
labels = 'Yes', 'No'
share = [percentage, 100 - percentage]

plt.pie(share,labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Likelihood of Toss determining Win')
plt.show()
```

```python
out_types = ['caught', 'bowled', 'lbw', 'caught and bowled', 'hit wicket', 'stumped']
df = dataset.loc[dataset['dismissal_kind'].isin(out_types)]
sub_data = df.loc[:, ['bowling_team','over','bowler', 'dismissal_kind', 'fielder']]
```

```python
wickets = sub_data['bowler'].value_counts().max()
bowler = sub_data['bowler'].value_counts().idxmax()
print(bowler,'took',wickets,'to become the highest IPL wicket taker')
```

```python
top5_bowlers = sub_data['bowler'].value_counts().head(5).keys().tolist()
wickets = sub_data['bowler'].value_counts().head(5).tolist()
fig=plt.gcf()
fig.set_size_inches(15, 7)
plt.bar(top5_bowlers, wickets, color='yellow')
plt.title('Top wicket takers')
plt.xlabel('Player')
plt.ylabel('Wickets')
plt.show()
```
