```python
import pandas as pd
data = pd.read_csv(r'D:\Projects\Ipl\Ipl-Kaggle-Dataset-Analysis\Player.csv')
data = data.dropna()
delivery = pd.read_csv(r'D:\Projects\Ipl\Ipl-Kaggle-Dataset-Analysis\deliveries.csv')
sub_data11 = delivery.loc[:, ['batsman', 'batsman_runs']]
sub_data = sub_data11.groupby('batsman', as_index = False).sum()
sub_data = sub_data.sort_values(by=['batsman_runs'], ascending = False)
batsman = sub_data['batsman'].tolist()
runs = sub_data['batsman_runs'].tolist()
```

```python
sub_data14 = delivery.loc[:, ['batsman', 'bowler', 'dismissal_kind']]
dismissal_kinds = ['caught',
 'bowled',
 'run out',
 'lbw',
 'caught and bowled',
 'stumped',
 'hit wicket']
is_bowler = (sub_data14['dismissal_kind'] == 'bowled') | (sub_data14['dismissal_kind'] == 'caught') | (sub_data14['dismissal_kind'] == 'caught and bowled') | (sub_data14['dismissal_kind'] == 'stumped') | (sub_data14['dismissal_kind'] == 'lbw') | (sub_data14['dismissal_kind'] == 'hit wicket')
bowler_data = sub_data14[is_bowler].loc[:, ['bowler']]
bowler = bowler_data['bowler'].value_counts().keys().tolist()
wickets = bowler_data['bowler'].value_counts().tolist()
```

```python
d1 = {'Player':bowler,'Wickets':wickets}
d2 = {'Player':batsman,'Runs':runs}
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
df1
```

```python
df2
```

```python
df = pd.merge(df2,
                 df1,
                 on='Player', how = 'left')
df.columns = ['Player', 'Runs', 'Wickets']
df = df.fillna(0)
```

```python
df
```

```python
from datetime import datetime
from datetime import date
def calculate_age(born):
    born = datetime.strptime(born, "%m/%d/%Y").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

data['Age'] = data['DOB'].apply(calculate_age)
data = data.loc[:, ['Player_Name','Age','Batting_Hand','Bowling_Skill','Country']]
```

```python
data.columns = ['Player', 'Age', 'Batting Skill', 'Bowling Skill', 'Country']
dataset = pd.merge(df, data, on = 'Player')
```

```python
dataset
```

```python
def label_player(row):
    if row['Runs'] > 600  and row['Wickets'] > 25:
        return 'All Rounder'
    elif row['Runs'] > 600:
        return 'Batsman'
    elif row['Wickets'] > 25:
        return 'Bowler'                              
    return 'Other'
```

```python
dataset['Type'] = dataset.apply(lambda row: label_player(row),axis=1)
```

```python
def rate_player(row):
    if row['Runs'] < 500:
        bat_value = 2
    elif row['Runs'] >= 500 and row['Runs'] < 1000:
        bat_value = 3
    elif row['Runs'] >= 1000 and row['Runs'] < 1500:
        bat_value = 4
    elif row['Runs'] >= 1500 and row['Runs'] < 2000:
        bat_value = 5
    elif row['Runs'] >= 2000 and row['Runs'] < 2500:
        bat_value = 6
    elif row['Runs'] >= 2500 and row['Runs'] < 3000:
        bat_value = 7
    elif row['Runs'] >= 3000 and row['Runs'] < 3500:
        bat_value = 8
    elif row['Runs'] >= 3500 and row['Runs'] < 4000:
        bat_value = 9
    elif row['Runs'] >= 4000:
        bat_value = 10
    if row['Wickets'] < 10:
        bowl_value = 2
    elif row['Wickets'] >= 10 and row['Wickets'] < 20:
        bowl_value = 3
    elif row['Wickets'] >= 20 and row['Wickets'] < 30:
        bowl_value = 4
    elif row['Wickets'] >= 30 and row['Wickets'] < 50:
        bowl_value = 5
    elif row['Wickets'] >= 50 and row['Wickets'] < 75:
        bowl_value = 6
    elif row['Wickets'] >= 75 and row['Wickets'] < 100:
        bowl_value = 7
    elif row['Wickets'] >= 100 and row['Wickets'] < 125:
        bowl_value = 8
    elif row['Wickets'] >= 125 and row['Wickets'] < 140:
        bowl_value = 9
    elif row['Wickets'] >= 140:
        bowl_value = 10
    if row['Age'] >= 35:
        age_value = 2
    elif row['Age'] >= 35 and row['Age'] < 40:
        age_value = 3
    elif row['Age'] >= 32 and row['Age'] < 35:
        age_value = 4
    elif row['Age'] >= 30 and row['Age'] < 32:
        age_value = 5
    elif row['Age'] >= 28 and row['Age'] < 30:
        age_value = 6
    elif row['Age'] >= 26 and row['Age'] < 28:
        age_value = 7
    elif row['Age'] >= 24 and row['Age'] < 26:
        age_value = 8
    elif row['Age'] >= 22 and row['Age'] < 24:
        age_value = 9
    elif row['Age'] <= 22:
        age_value = 10
    return round(((bat_value + bowl_value + age_value)/3),2)
```

```python
dataset['Rating'] = dataset.apply(lambda row: rate_player(row),axis=1) 
```

```python
dataset = dataset.loc[:,['Player', 'Runs', 'Wickets', 'Age', 'Batting Skill','Bowling Skill', 'Country', 'Type', 'Rating']]
max_val = dataset['Rating'].max()
min_val = dataset['Rating'].min()
```

```python
values = dataset['Rating'].tolist()
scaled_ratings = [round(((((10.0-0.0)*(x-min_val))/(max_val-min_val))+0.0),2) for x in values]
dataset['Rating'] = scaled_ratings
```

```python
def buy_type(row):
    if row['Rating'] < 2.5:
        return 'Avoid'
    elif row['Rating'] >= 2.5 and row['Rating'] < 5.0:
        return 'Maybe'
    elif row['Rating'] >= 5.0 and row['Rating'] < 7.5:
        return 'Good Buy'
    else:
        return 'Excellent Purchase'
```

```python
dataset['Buy type'] = dataset.apply(lambda row: buy_type(row),axis=1) 
```

```python
young_star = (dataset['Age'] < 28) & (dataset['Buy type'] == 'Good Buy')
data = dataset[young_star]
data
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X = dataset.iloc[:, 1:4].values
Y = dataset.iloc[:,9].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state = 40)

model = XGBClassifier()

model.fit(X_train, Y_train)

#Make predictions for test data
y_pred = model.predict(X_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
print (cm)
from sklearn.metrics import accuracy_score
print(round(accuracy_score(Y_test, y_pred),2)*100,"% Accuracy")
```
